```python



# Define a function to convert 'X Lakhs', 'X Crore', and remove dollar symbol
def convert_amount(value):
    if isinstance(value, str):
        # Remove commas for numeric conversion
        value = value.replace(',', '')
        
        # Handle Lakhs
        if 'Lakh' in value:
            number_str = value.replace('₹', '').replace('Lakh', '').strip()
            number = float(number_str)
            return int(number * 100000)  # 1 Lakh = 100,000
        # Handle Crores
        elif 'Crore' in value:
            number_str = value.replace('₹', '').replace('Crore', '').strip()
            number = float(number_str)
            return int(number * 10000000)  # 1 Crore = 10,000,000
        # Handle Dollar values
        elif '₹' in value:
            number_str = value.replace('₹', '').strip()
            return int(number_str)  # Convert to integer
        else:
            return int(value)  # Direct conversion for plain integers
    return value

# Apply the conversion function to the 'Amount' column
df['price'] = df['price'].apply(convert_amount)

df

```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import Ridge, Lasso, LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
# statistical details of the data
file
```

```python
# create x and y datas

x=file.drop(columns=["price"],axis=1)
y=file["price"]
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
x_scaled=scaler.fit_transform(x)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
# Initialize models
models = {
    'Linear Regression': LinearRegression(),
    'Decision Tree': DecisionTreeRegressor(),
    'Ridge Regression': Ridge(alpha=1.0),
    'Lasso Regression': Lasso(alpha=0.1),
    'Random Forest': RandomForestRegressor(),
    'Gradient Boosting': GradientBoostingRegressor(),
    'K-Nearest Neighbors': KNeighborsRegressor()
}

# Train and evaluate each model
results = {}

for model_name, model in models.items():
    # Fit the model
    model.fit(X_train, y_train)

    # Make predictions
    y_pred = model.predict(X_test)

    # Calculate metrics
    mae = mean_absolute_error(y_test, y_pred)
    mse = mean_squared_error(y_test, y_pred)
    rmse = mean_squared_error(y_test, y_pred, squared=False)  # RMSE
    r2 = r2_score(y_test, y_pred)  # R-squared

    results[model_name] = {
        'MAE': mae,
        'MSE': mse,
        'RMSE': rmse,
        'R²': r2
    }

# Display results
results_df = pd.DataFrame(results).T
print(results_df)
```

Hyperparameter tuning

```python
# parameters for model
from sklearn.model_selection import RandomizedSearchCV

param_grid = {
    'n_estimators' : [100, 150, 200, 300],
    'max_depth': [10, 20, 30, None],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4],
    'max_features': ['sqrt', 'log2', None],

}
```

```python
# tuning the model using grid search cv
grid_tuning=RandomizedSearchCV(estimator=RandomForestRegressor(),
                            param_distributions=param_grid, cv=5,verbose = 2, n_jobs=-1)


grid_tuning.fit(X_train,y_train)
```

```python
best_params = grid_tuning.best_params_
best_score = grid_tuning.cv_results_['mean_test_score'][grid_tuning.best_index_]

print(best_params, best_score)
```

```python
import numpy as np
# final model fit and prediction
final_model=RandomForestRegressor(n_estimators=200,max_features='sqrt',max_depth=30,min_samples_leaf=1,
                                  min_samples_split=2,random_state=42)

final_model.fit(X_train,y_train)

#Model evaluation
y_pred_fm=final_model.predict(X_test)
MSE_fm=mean_squared_error(y_test,y_pred_fm)
MAE_fm=mean_absolute_error(y_test,y_pred_fm)
RMSE_fm=np.sqrt(mean_squared_error(y_test,y_pred_fm))
r_squr_fm=r2_score(y_test,y_pred_fm)

print(f"Trainscore: {final_model.score(X_train, y_train)}")
print(f"Testscore: {final_model.score(X_test, y_test)}")

# create dataframe to view above evaluation metrics
rf_1=pd.DataFrame({"model":"RandomForestRegressor","MAE":[MAE_fm],"MSE":[MSE_fm],"RMSE":[RMSE_fm],"R2":r_squr_fm})
rf_1
```
